#### Insuring all Club URLS are for year 2022

```python
List_of_club_urls = [s.replace('2021', '2022') for s in List_of_club_urls]
```

#### Looping through the URLS and Scraping the needed Data in a Dataframe

```python
data = pd.DataFrame(columns= ['Player','Team', 'Date of birth / Age', 'Height', 'Joined', 'Contract', 'Market value'])
for club in List_of_club_urls:
    ssl._create_default_https_context = ssl._create_unverified_context
    from urllib.request import Request, urlopen    
    url = f"{club}"    
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})    
    html = urlopen(req)

    scraped = pd.read_html(html)
    df = scraped[1]
    df = df[['Player', 'Date of birth / Age', 'Height', 'Joined', 'Contract', 'Market value']]
    df = df.dropna()
    two = df.Player.str.split(pat=' ',expand=True)
    two['first']= two.iloc[:, 0] + ' ' + two.iloc[:, 1]
    df['Player'] = two['first']
    df['Team'] = club.split('/')[-8].replace("-", " ")
    data = pd.concat([data, df], axis=0)
data
```

```python
data.index = range(len(data.index))
data
```

```python
data.to_excel("TransferMarket for Germany's 4th tier players.xlsx", encoding='utf-8', index=False)
```
